```python
import time
import logging
from datetime import datetime as dt
from IPython import get_ipython
import traceback

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Get the current IPython instance
ipython = get_ipython()

# Path to your notebook
notebook_path = "./simplied_daily_fois_alert.ipynb"  # Adjust this path to the location of your notebook

if ipython:
    while True:
        try:
            logging.info(f'Starting notebook at {dt.now()}')
            ipython.magic(f"%run {notebook_path}")
            logging.info(f'Completed notebook at {dt.now()}')
            time.sleep(3600)  # Run every hour
        except Exception as e:
            logging.error(f'Error occurred: {e}')
            logging.error(traceback.format_exc())
else:
    logging.error("This script must be run in a Jupyter environment.")
```
